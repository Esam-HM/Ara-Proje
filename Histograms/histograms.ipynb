```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
furkan_excel_path = "./dataset/Furkan Excel.xlsx"
gamze_excel_path = "./dataset/Gamze Excel.xlsx"
total_excel_path = "./dataset/Total Excel.xlsx"
```

```python
furkan_labels = pd.read_excel(furkan_excel_path,header=None,keep_default_na=False)
total_labels = pd.read_excel(total_excel_path,header=None,keep_default_na=False)
gamze_labels = pd.read_excel(gamze_excel_path,header=None,keep_default_na=False)
print(len(furkan_labels),len(gamze_labels),len(total_labels))
```

```python
classes = ["Hatasız","Hatalı"]
error_types = ["Baş Aşağı","Baş Yukarı","Çentiğin Önden Isırılması","Çentiğin Arkadan Isırılması",
              "Hasta Başını Sağa Sola Çevirmiş","Hasta Başını Sağa Sola Yatırmış"]
```

```python
def create_histogram(classes:list,excel_file:list):
    values = [0]*len(classes)
    if len(classes)>2:
        for i in range(0,len(excel_file)):
            id = excel_file.iloc[i,2]
            if id !='' and id !=0:
                values[id-1]+=1
        colors = ["#9467bd","#17becf","#e377c2","#2ca02c","#ff7f0e","#bcbd22"]
        labels = ["1","2","3","4","5","6"]                  ## isimler x eksenine sığmadı.
        title = "Hata tipleri ve sayıları"
    else:
        for i in range(0,len(excel_file)):
            values[excel_file.iloc[i,1]]+=1
        colors = ["#2ca02c","#d62728"]
        labels = ["0","1"]
        title = "Hatalı ve Hatasız görüntülerin sayısı"
    plt.figure(figsize=(7,5))
    plt.bar(labels,values,color = colors,label = classes)
    for i in range(0,len(classes)):
        plt.text(i,values[i]//2,values[i],ha='center',fontsize = 12)            ## Her sutünün görüntü sayısı yazdırma
    plt.legend(loc="upper right")
    plt.title(title)
    plt.xlabel("Sınıf")
    plt.ylabel("Görüntü sayısı")
    plt.tight_layout()
    plt.show()
```

## Data Setin Tamamı

```python
create_histogram(classes,total_labels)
create_histogram(error_types,total_labels)
```

## Furkan Pan Data Seti

```python
create_histogram(classes,furkan_labels)
create_histogram(error_types,furkan_labels)
```

## Gamze Pan Dataseti

```python
create_histogram(classes,gamze_labels)
create_histogram(error_types,gamze_labels)
```
